```python
import numpy as np
import matplotlib.pyplot as plt

# Define Solarized color scheme
base02 = "#073642"
base01 = "#586e75"
base00 = "#657b83"
yellow = "#b58900"
orange = "#cb4b16"
red = "#dc322f"
blue = "#268bd2"
green = "#859900"

# Set up the figure
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Single Point Calibration
ax = axes[0]
measured_param = np.linspace(0, 10, 100)
ideal_response = measured_param
actual_response = measured_param + 2  # Offset by 2

ax.plot(measured_param, ideal_response, label="Ideal Response", color=base00, linestyle="--")
ax.plot(measured_param, actual_response, label="Actual Response (Offset Error)", color=green)
ax.scatter([5], [7], color=blue, label="Calibration Point")

# Add vertical distance indicator for offset
ax.annotate("Offset", xy=(5, 7), xytext=(4.5, 4.5),
            arrowprops=dict(arrowstyle="<->", color=blue), fontsize=10, color=blue)


ax.set_title("Single Point Calibration", fontsize=14, color=base02)
ax.set_xlabel("Actual 'True' Value", fontsize=12, color=base02)
ax.set_ylabel("Measured Value (With Error)", fontsize=12, color=base02)
ax.tick_params(colors=base02)
ax.legend(loc="upper left", fontsize=10, frameon=False)
ax.spines['top'].set_color(base01)
ax.spines['right'].set_color(base01)
ax.spines['bottom'].set_color(base01)
ax.spines['left'].set_color(base01)

# Two Point Calibration
ax = axes[1]
raw_low = 2
raw_high = 8
reference_low = 1
reference_high = 10
actual_response = (measured_param - raw_low) / (raw_high - raw_low) * (reference_high - reference_low) + reference_low

ax.plot(measured_param, ideal_response, label="Ideal Response", color=base00, linestyle="--")
ax.plot(measured_param, actual_response, label="Actual Response", color=red)
ax.scatter([raw_low, raw_high], [reference_low, reference_high], color=blue, label="Calibration Points")
ax.annotate("Reference Low", xy=(raw_low, reference_low), xytext=(raw_low - 1, reference_low + 1),
            arrowprops=dict(facecolor=blue, shrink=0.05), fontsize=10, color=blue)
ax.annotate("Reference High", xy=(raw_high, reference_high), xytext=(raw_high + 1, reference_high - 1),
            arrowprops=dict(facecolor=blue, shrink=0.05), fontsize=10, color=blue)

ax.set_title("Two Point Calibration", fontsize=14, color=base02)
ax.set_ylabel("Measured Value (With Error)", fontsize=12, color=base02)
ax.set_xlabel("Actual 'True' Value", fontsize=12, color=base02)
ax.tick_params(colors=base02)
ax.legend(loc="upper left", fontsize=10, frameon=False)
ax.spines['top'].set_color(base01)
ax.spines['right'].set_color(base01)
ax.spines['bottom'].set_color(base01)
ax.spines['left'].set_color(base01)

# Multi Point Calibration
ax = axes[2]
actual_response = measured_param + 0.1 * measured_param**2  # Non-linear response
calibration_points_x = [2, 5, 8]
calibration_points_y = calibration_points_x + 0.1 * np.array(calibration_points_x)**2

ax.plot(measured_param, ideal_response, label="Ideal Response", color=base00, linestyle="--")
ax.plot(measured_param, actual_response, label="Actual Response (Non-Linearity Error)", color=red)
ax.scatter(calibration_points_x, calibration_points_y, color=blue, label="Calibration Points")


ax.set_title("Multi Point Calibration", fontsize=14, color=base02)
ax.set_ylabel("Measured Value (With Error)", fontsize=12, color=base02)
ax.set_xlabel("Actual 'True' Value", fontsize=12, color=base02)
ax.tick_params(colors=base02)
ax.legend(loc="upper left", fontsize=10, frameon=False)
ax.spines['top'].set_color(base01)
ax.spines['right'].set_color(base01)
ax.spines['bottom'].set_color(base01)
ax.spines['left'].set_color(base01)

# Adjust layout and background
fig.patch.set_facecolor("#eee8d5")  # Base2
plt.tight_layout()
plt.show()

```
