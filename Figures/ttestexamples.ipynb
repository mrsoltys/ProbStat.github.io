```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Solarized color scheme
solarized_base2 = "#eee8d5"
solarized_base02 = "#073642"
solarized_blue = "#268bd2"
solarized_orange = "#cb4b16"
solarized_violet = "#6c71c4"

# Define parameters
df = 9  # degrees of freedom
alpha = 0.05  # significance level
t_critical = stats.t.ppf(alpha, df)  # left-tailed critical value
x = np.linspace(-4, 4, 1000)  # x range for plotting
y = stats.t.pdf(x, df)  # t-distribution PDF

# Create figure
fig, ax = plt.subplots(figsize=(8, 5), facecolor=solarized_base2)
ax.set_facecolor(solarized_base2)

# Plot t-distribution
ax.plot(x, y, color=solarized_violet, linewidth=2, label="t-distribution (df=9)")

# Shade rejection region (left-tailed)
x_fill = np.linspace(-4, t_critical, 300)
y_fill = stats.t.pdf(x_fill, df)
ax.fill_between(x_fill, y_fill, color=solarized_orange, alpha=0.7, label="Rejection Region ($\\alpha=0.05$)")

# Mark critical t-value
ax.axvline(t_critical, color=solarized_blue, linestyle="dashed", linewidth=2, label=f"Critical Value: t = {t_critical:.3f}")

# Labels and title
ax.set_title("One-Tailed t-Test: Left-Tailed Rejection Region", fontsize=12, color=solarized_base02)
ax.set_xlabel("t-value", fontsize=10, color=solarized_base02)
ax.set_ylabel("Probability Density", fontsize=10, color=solarized_base02)
ax.tick_params(colors=solarized_base02)

# Legend
ax.legend(frameon=False, fontsize=10)

# Show plot
plt.show()
fig.savefig("oneTailedT-test.png", format="png", dpi=300, bbox_inches="tight")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Solarized color scheme
solarized_base2 = "#eee8d5"
solarized_base02 = "#073642"
solarized_blue = "#268bd2"
solarized_orange = "#cb4b16"
solarized_violet = "#6c71c4"

# Define parameters
df = 14  # degrees of freedom
alpha = 0.08  # significance level
t_critical = stats.t.ppf(1 - alpha / 2, df)  # two-tailed critical value
x = np.linspace(-4, 4, 1000)  # x range for plotting
y = stats.t.pdf(x, df)  # t-distribution PDF

# Create figure
fig, ax = plt.subplots(figsize=(8, 5), facecolor=solarized_base2)
ax.set_facecolor(solarized_base2)

# Plot t-distribution
ax.plot(x, y, color=solarized_violet, linewidth=2, label="t-distribution (df=14)")

# Shade rejection region (two-tailed)
x_fill_left = np.linspace(-4, -t_critical, 300)
y_fill_left = stats.t.pdf(x_fill_left, df)
ax.fill_between(x_fill_left, y_fill_left, color=solarized_orange, alpha=0.7)

x_fill_right = np.linspace(t_critical, 4, 300)
y_fill_right = stats.t.pdf(x_fill_right, df)
ax.fill_between(x_fill_right, y_fill_right, color=solarized_orange, alpha=0.7, label="Rejection Region ($\\alpha=0.08$)")

# Mark critical t-values
ax.axvline(-t_critical, color=solarized_blue, linestyle="dashed", linewidth=2, label=f"Critical Value: t = {t_critical:.3f}")
ax.axvline(t_critical, color=solarized_blue, linestyle="dashed", linewidth=2)

# Labels and title
ax.set_title("Two-Tailed t-Test: Rejection Regions", fontsize=12, color=solarized_base02)
ax.set_xlabel("t-value", fontsize=10, color=solarized_base02)
ax.set_ylabel("Probability Density", fontsize=10, color=solarized_base02)
ax.tick_params(colors=solarized_base02)

# Legend
ax.legend(frameon=False, fontsize=10)

# Show plot
plt.show()
fig.savefig("twoTailedT-test.png", format="png", dpi=300, bbox_inches="tight")
```
