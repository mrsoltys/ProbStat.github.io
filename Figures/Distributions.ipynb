```python
import numpy as np
import matplotlib.pyplot as plt

# Define Solarized color scheme
base02 = "#073642"
base01 = "#586e75"
base00 = "#657b83"
blue = "#268bd2"

# Initialize the figure
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.flatten()

# Define new distributions
x = np.linspace(-3, 3, 500)
x_positive = np.linspace(0, 3, 500)
distributions = {
    "Normal (np.random.normal)": np.exp(-x**2 / 2) / np.sqrt(2 * np.pi),
    "Negatively skewed (custom)": np.exp(-(x + 1)**2 / 2) / np.sqrt(2 * np.pi),
    "Positively skewed (custom)": np.exp(-(x - 1)**2 / 2) / np.sqrt(2 * np.pi),
    "Bimodal (mixture)": 0.5 * np.exp(-x**2 / 2) / np.sqrt(2 * np.pi) + 0.5 * np.exp(-(x - 2)**2 / 2) / np.sqrt(2 * np.pi),
    "Exponential (np.random.exponential)": np.exp(-x_positive),
    "Chi-squared (np.random.chisquare)": x_positive * np.exp(-x_positive),
    "Uniform (np.random.uniform)": np.ones_like(x) * 0.3,
    "Laplace (np.random.laplace)": 0.5 * np.exp(-np.abs(x)),
    "Gamma (np.random.gamma)": (x_positive**2) * np.exp(-x_positive),
}

# Titles and x-values
titles = list(distributions.keys())
x_values = [x, x, x, x, x_positive, x_positive, x, x, x_positive]

# Plot each distribution
for ax, title, dist_x, y in zip(axes, titles, x_values, distributions.values()):
    ax.plot(dist_x, y, color=blue, linewidth=4)
    ax.set_title(title, fontsize=14, color=base02)
    ax.set_yticks([])
    ax.set_xticks([])
    ax.spines['top'].set_color(base01)
    ax.spines['right'].set_color(base01)
    ax.spines['bottom'].set_color(base01)
    ax.spines['left'].set_color(base01)
    ax.tick_params(colors=base02)
    ax.set_facecolor("#eee8d5")

# Add a main title
fig.suptitle("Common Distributions", fontsize=20, color=base02)

# Adjust layout and background
fig.tight_layout(rect=[0, 0, 1, 0.95])
fig.patch.set_facecolor("#eee8d5")
plt.show()
fig.savefig("Distributions.png", format="png", dpi=300, bbox_inches="tight")

```
