```python
# Re-import libraries and redefine variables due to state reset
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Solarized color scheme
base03 = "#002b36"
base02 = "#073642"
base01 = "#586e75"
base00 = "#657b83"
base0 = "#839496"
base1 = "#93a1a1"
yellow = "#b58900"
orange = "#cb4b16"
red = "#dc322f"
magenta = "#d33682"
violet = "#6c71c4"
blue = "#268bd2"
cyan = "#2aa198"
green = "#859900"
base2 = "#eee8d5"

# Mean and standard deviation for the normal distribution
mean = 0
std_dev = 1

# Generate x values
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mean, std_dev)

# Create the figure and the plot with updated labels and styles
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, pdf, color=blue, linewidth=2)

# Fill areas under the curve for 68-95-99.7 rule
ax.fill_between(x, 0, pdf, where=(x >= mean - std_dev) & (x <= mean + std_dev), color=yellow, alpha=0.7)
ax.fill_between(x, 0, pdf, where=(x >= mean - 2*std_dev) & (x <= mean + 2*std_dev), color=orange, alpha=0.5)
ax.fill_between(x, 0, pdf, where=(x >= mean - 3*std_dev) & (x <= mean + 3*std_dev), color=red, alpha=0.3)

# Adding vertical lines for mean and standard deviations
for i, color in zip(range(-3, 4), [green, cyan, violet, blue, violet, cyan, green]):
    ax.axvline(mean + i*std_dev, color=color, linestyle="dashed", linewidth=1.5)

# Adding horizontal bars for 68%, 95%, and 99.7% (flipped order)
ax.hlines(y=0.24, xmin=mean - std_dev, xmax=mean + std_dev, color=yellow, linewidth=2)
ax.hlines(y=0.055, xmin=mean - 2*std_dev, xmax=mean + 2*std_dev, color=orange, linewidth=2)
ax.hlines(y=0.005, xmin=mean - 3*std_dev, xmax=mean + 3*std_dev, color=red, linewidth=2)

# Annotating the horizontal bars
ax.text(mean, 0.25, "68%", color=base02, fontsize=12, ha="center")
ax.text(mean, 0.065, "95%", color=base02, fontsize=12, ha="center")
ax.text(mean, 0.015, "99.7%", color=base02, fontsize=12, ha="center")


# Update x-axis with custom labels
x_ticks = [-3, -2, -1, 0, 1, 2, 3]
x_labels = [
    "$\mu - 3\sigma$", "$\mu - 2\sigma$", "$\mu - \sigma$",
    "$\mu$", "$\mu + \sigma$", "$\mu + 2\sigma$", "$\mu + 3\sigma$"
]
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels, fontsize=12, color=base02)

# Labeling
ax.set_title("68-95-99.7 Rule", fontsize=16, color=base02)
ax.set_xlabel("", fontsize=14, color=base02)  # No x-label, as labels are on the ticks
ax.set_ylabel("Probability Density", fontsize=14, color=base02)
ax.tick_params(axis='y', colors=base02)
ax.spines['top'].set_color(base01)
ax.spines['right'].set_color(base01)
ax.spines['bottom'].set_color(base01)
ax.spines['left'].set_color(base01)

# Background color adjustments
ax.set_facecolor(base2)
fig.patch.set_facecolor(base2)

# Show the plot
plt.show()
fig.savefig("68-95-99.7-rule.png", format="png", dpi=300, bbox_inches="tight")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Generate the data (non-symmetric distribution using lognormal)
data = np.random.lognormal(mean=0, sigma=0.5, size=1000)

# Calculate boxplot statistics
q1 = np.percentile(data, 25)
median = np.percentile(data, 50)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr
mean = np.mean(data)
std_dev = np.std(data)

# Define the Solarized color scheme
base02 = "#073642"
base01 = "#586e75"
base00 = "#657b83"
yellow = "#b58900"
orange = "#cb4b16"
red = "#dc322f"
blue = "#268bd2"
violet = "#6c71c4"

# Create the figure with two subplots: one for the histogram and one for the boxplot
fig, (ax_hist, ax_box) = plt.subplots(2, 1, figsize=(12, 8), gridspec_kw={'height_ratios': [2, 1]})

# Histogram of the data (top plot)
ax_hist.hist(data, bins=30, color=orange, alpha=0.6, edgecolor="black")
ax_hist.axvline(mean, color=blue, linestyle="dashed", linewidth=1.5, label="Mean")
ax_hist.axvline(median, color=red, linewidth=2, label="Median")
ax_hist.axvline(mean - std_dev, color=violet, linestyle="dotted", linewidth=1.5)
ax_hist.axvline(mean + std_dev, color=violet, linestyle="dotted", linewidth=1.5)
ax_hist.set_title("Histogram of Data", fontsize=14, color=base02)
ax_hist.set_ylabel("Frequency", fontsize=12, color=base02)
ax_hist.tick_params(axis='x', colors=base02)
ax_hist.tick_params(axis='y', colors=base02)
ax_hist.spines['top'].set_color(base01)
ax_hist.spines['right'].set_color(base01)
ax_hist.spines['bottom'].set_color(base01)
ax_hist.spines['left'].set_color(base01)
ax_hist.text(mean - std_dev, 180, "-1", ha="center", fontsize=10, color=blue)
ax_hist.text(mean + std_dev, 180, "+1", ha="center", fontsize=10, color=blue)
ax_hist.text(mean, 180, "Mean", ha="center", fontsize=10, color=blue)

# Align x-axis of the histogram and boxplot
ax_hist.set_xlim(mean - 3 * std_dev, mean + 3 * std_dev)
ax_hist.set_xticks([mean - 3 * std_dev, mean - 2 * std_dev, mean - std_dev, mean, mean + std_dev, mean + 2 * std_dev, mean + 3 * std_dev])
ax_hist.set_xticklabels([])  # Remove x-labels from the histogram

# Boxplot (bottom plot)
ax_box.add_patch(plt.Rectangle((q1, 0.45), iqr, 0.1, color=orange, alpha=0.6))
ax_box.plot([median, median], [0.45, 0.55], color=red, linewidth=4)
ax_box.plot([lower_whisker, q1], [0.5, 0.5], color=violet, linewidth=1.5)
ax_box.plot([q3, upper_whisker], [0.5, 0.5], color=violet, linewidth=1.5)
ax_box.plot([lower_whisker, lower_whisker], [0.47, 0.53], color=violet, linestyle="solid", linewidth=1.5)
ax_box.plot([upper_whisker, upper_whisker], [0.47, 0.53], color=violet, linestyle="solid", linewidth=1.5)

# Draw outliers
outliers = data[(data < lower_whisker) | (data > upper_whisker)]
ax_box.scatter(outliers, [0.5] * len(outliers), color=yellow, s=30, zorder=5)

# Draw standard deviations (without mean line)
ax_box.plot([mean - std_dev, mean + std_dev], [0.5, 0.5], color=blue, linestyle="solid", linewidth=1)

# Annotations for the boxplot
ax_box.text(q1, 0.43, "Q1", ha="center", fontsize=10, color=base02)
ax_box.text(q3, 0.43, "Q3", ha="center", fontsize=10, color=base02)
ax_box.text(median, 0.41, "Median\n(Q2, 50th percentile)", ha="center", fontsize=10, color=red)
ax_box.text(lower_whisker, 0.55, "Min (Q1 - 1.5*IQR)", ha="center", fontsize=10, color=violet)
ax_box.text(upper_whisker, 0.55, "Max (Q3 + 1.5*IQR)", ha="center", fontsize=10, color=violet)
ax_box.text(upper_whisker+.21, 0.47, "Outliers", ha="center", fontsize=10, color=yellow)

ax_box.hlines(y=0.57, xmin=q1, xmax=q3, color=orange, linewidth=2)

# Annotating the horizontal bars
ax_box.text(median, 0.58, "IQR", color=orange, fontsize=12, ha="center")


# Set x-axis limits and ticks for boxplot
ax_box.set_xlim(mean - 3 * std_dev, mean + 3 * std_dev)
ax_box.set_xticks([mean - 3 * std_dev, mean - 2 * std_dev, mean - std_dev, mean, mean + std_dev, mean + 2 * std_dev, mean + 3 * std_dev])
ax_box.set_xticklabels([
    "$\mu - 3\sigma$", "$\mu - 2\sigma$", "$\mu - \sigma$",
    "$\mu$", "$\mu + \sigma$", "$\mu + 2\sigma$", "$\mu + 3\sigma$"
], fontsize=10, color=base02)

# Remove y-axis for boxplot
ax_box.set_yticks([])
ax_box.set_ylim(.4,.6)
ax_box.set_ylabel("")

# Add title for boxplot
ax_box.set_title("Boxplot Summarizing the Data", fontsize=14, color=base02)

# Adjust background colors for both plots
ax_hist.set_facecolor("#eee8d5")  # Base2
ax_box.set_facecolor("#eee8d5")
fig.patch.set_facecolor("#eee8d5")

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
fig.savefig("BoxPlot.png", format="png", dpi=300, bbox_inches="tight")


```
